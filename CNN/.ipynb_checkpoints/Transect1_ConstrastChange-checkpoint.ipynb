```python
import sys

sys.path.append('./utils/')
import numpy as np
import cv2
import gdal
import loadcoraldata_utils as coralutils
from matplotlib import pyplot as plt

%matplotlib notebook
```

```python
# Test export of patches from map
transect1_path = '../Images/Transect 1 Hi-Res.tiff'
transect1_truth_path = '../Images/Transect 1 Truth data.tif'

image_size = 25
labels = ('Sand', 'Branching', 'Mounding', 'Rock')
Transect1 = coralutils.CoralData(transect1_path, Truthpath=transect1_truth_path, truth_key=[16,160,198,38], load_type="cv2")
Transect1_bgr = Transect1.image[:,:,:-1]
Transect1_hsv = cv2.cvtColor(Transect1.image, cv2.COLOR_BGR2HSV)
Transect1_h = Transect1_hsv[:,:,0] + 15
Transect1_v = Transect1_hsv[:,:,2] + 25
Transect1_h[Transect1_h > 255] = 255
Transect1_v[Transect1_v > 255] = 255
Transect1_hsv[:,:,0] = Transect1_h
Transect1_hsv[:,:,2] = Transect1_v

# Transect1_hsv[:,:,2] = [[max(pixel - 25, 0) if pixel < 190 else min(pixel + 25, 255) for pixel in row] for row in Transect1_hsv[:,:,2]]

plt.figure()
plt.subplot(1,2,1)
plt.imshow(Transect1_bgr[...,::-1])
plt.subplot(1,2,2)
plt.imshow(cv2.cvtColor(Transect1_hsv, cv2.COLOR_HSV2RGB))
plt.show()
print(Transect1.image.shape)

cv2.imwrite('../Images/Transect 1 Hi-Res Skew.tiff', cv2.cvtColor(Transect1_hsv,cv2.COLOR_HSV2BGR))
```

```python
transect1_skew_path = '../Images/Transect 1 Hi-Res Skew.tiff'
output_trainpath = '../Images/Training_Patches_Skew/'
output_trainrefpath = '../Images/TrainingRef_Patches_Skew/'
output_trainfilename = 'NeMO_train.txt'
output_validpath = '../Images/Valid_Patches_Skew/'
output_validrefpath = '../Images/ValidRef_Patches_Skew/'
output_validfilename = 'NeMO_valid.txt'

Transect1_skew = coralutils.CoralData(transect1_skew_path, Truthpath=transect1_truth_path, truth_key=[16,160,198,38], load_type="cv2")
print("Outputing Training...")
Transect1_skew.export_segmentation_map(output_trainpath, output_trainrefpath, output_trainfilename, image_size = image_size, N = 2000, lastchannelremove = False, labelkey = labels, subdir=True)
print("\n")
print("Outputing Validation...")
Transect1_skew.export_segmentation_map(output_validpath, output_validrefpath, output_validfilename, image_size = image_size, N = 100, lastchannelremove = False, labelkey = labels, subdir=True)
```
